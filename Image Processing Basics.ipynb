```python
#Reading web cam
#cap = cv2.VideoCapture(0)

#Reading Video

import cv2
cap = cv2.VideoCapture('lane.mkv')

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        cv2.imshow('Frame',frame)
        
    else:
        break
    #waiting for 30 milisecons between frames and if K press, it will break
    
    #for any key
    #if cv2.waitKey(1)! = -1:
    if cv2.waitKey(30)==ord('k'):  
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread('car.jpg')

#draw a Line
cv2.line(img, (100,100) , (300,255), (255,255,0), 2)

#draw a retangle
cv2.rectangle(img, (50,50), (100,600), (255,0,0),3)

#for filled rectangle 
#thickness = -1

#draw a circle
cv2.circle(img, (300,150), 50, (255,255,0), -1)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#add text to image

import cv2


img = cv2.imread('car.jpg')

cv2.putText(img, 'Workshop', (0,130), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0), 2)
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#IMAGE ADDITION

import cv2
import numpy as np

img1 = cv2.imread('img1.jpg')
img2 = cv2.imread('img2.jpg')

add = cv2.add(img1, img2)
cv2.imshow('add', add)
cv2.imshow('Image', add)

#weighted addition

add_weighted = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
cv2.imshow('Image', add_weighted)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Threshold

#make an image readable using threshold

import cv2

img = cv2.imread('bookpage.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


ret,threshold = cv2.threshold(gray, 7, 255, cv2.THRESH_BINARY)
#cv2.imshow('original', img)
#cv2.imshow('gray', gray)
cv2.imshow('Threshold', threshold)

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
##############################################

#Adding image to image

import cv2

logo = cv2.imread('logo.jpg')
img = cv2.imread('car.jpg')

rows, cols, channels = logo.shape

#roi = region of image
roi = img[0:rows, 0:cols]

logo_gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)
#cv2.imshow('Gray_Logo', logo_gray)

ret, mask = cv2.threshold(logo_gray, 10,255, cv2.THRESH_BINARY)


mask_inv = cv2.bitwise_not(mask) # inversion of mask operation

fg = cv2.bitwise_and(logo, logo, mask=mask)
bg = cv2.bitwise_and(roi, roi, mask=mask_inv)

dst = cv2.add(fg,bg)

img[0:rows, 0:cols] = dst

#cv2.imshow('Gray_Logo', logo_gray)
#cv2.imshow('MASK', mask)
#cv2.imshow('INV_MASK', mask_inv) 
#cv2.imshow('fg',fg)
#cv2.imshow('bg',bg)
#cv2.imshow('dst',dst)
cv2.imshow('result', img)

cv2.waitKey(0)
cv2.destroyAllWindows()


```
